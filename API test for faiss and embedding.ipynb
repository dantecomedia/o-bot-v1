```python
import requests
import json

url = "http://ec2-52-66-157-239.ap-south-1.compute.amazonaws.com:5000/generate_embeddings"

payload = {
    "context": {
        "q_1": "What is a Machine Learning Engineer?",
        "follow_up": [
            "What are the skills required for a Machine Learning Engineer?",
            "What is the average salary for a Machine Learning Engineer?"
        ]
    }
}
headers = {
    'Content-Type': 'application/json'
}

response = requests.post(url, headers=headers, data=json.dumps(payload))

print(response.status_code)
print(response.json())
```
